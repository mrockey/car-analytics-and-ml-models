## Performing Cleaning of the Scraped Cars.com Data

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#Loading scrapped Data
df = pd.read_pickle('/dsa/groups/casestudy2022su/team05/carscom_v01.pkl') 
```

```python
df.info()
df.head()
```

### Separating Vehicle into Year, Make, and Model

```python
# Splitting Vehcile column into Year and Make
new = df["Vehicle"].str.split(" ", n = 1, expand = True)

#Assigning the split to new columns.
df["Year"]=new[0]
df["Make"]=new[1]

#Dropping Vehicle column
df.drop(columns = ["Vehicle"], inplace = True)
```

```python
df.head()
```

```python
#Splitting the Make into two columns to create a Model column
new2 = df["Make"].str.split(" ", n = 1, expand = True)

#Assigning
df["Make2"]=new2[0]
df["Model"]=new2[1]

#Dropping Make
df.drop(columns = ["Make"], inplace = True)
```

```python
df.head()
```

```python
#Reordering Columns
column_to_reorder = df.pop('Model')
column_to_reorder2 = df.pop('Make2')
column_to_reorder3 = df.pop('Year')
df.insert(0, 'Model', column_to_reorder)
df.insert(0, 'Make2', column_to_reorder2)
df.insert(0, 'Year', column_to_reorder3)
```

```python
# Changing the name of Make2 back to Make
df = df.rename(columns={"Make2": "Make"})
```

```python
df.head()
```

### Limiting the models toat least 15 or more unique value counts.

```python
df.groupby('Make')['Model'].value_counts().sort_values(ascending=False)
```

```python
#Limiting to Models with unique values greaterh than 14.
#Reduction of total models from 8422 to 2643
df['Model'].value_counts().loc[lambda x : x>14]
```

```python
#Limiting to Models with unique values greater than 14 and storing in a list called ModelsToKeep
ModelsToKeep = df['Model'].value_counts().loc[lambda x : x>14].index.tolist()
```

```python
ModelsToKeep
```

```python
#Using .isin to keep only models listed in ModelsToKeep. Reseting the index...
df2 = df[df['Model'].isin(ModelsToKeep)].reset_index()
```

```python
#Reduction of total instances from 420916 to 400741
df2.info()
df2.head()
```

### Clean the Mileage, Price and Deal_Rating Columns

```python
#Exract the mileage column and split the string to remove the characters.
mileage = df2['Mileage']
mileage_column = pd.DataFrame(mileage.str.split(expand = True)).drop(columns=[1])
mileage_column.columns = ['Mileage']
mileage_column['Mileage'] = mileage_column['Mileage'].str.replace(',', '')
mileage_column.head(10)
```

```python
#Exract the price column and split the string to remove the characters.
price = df2['Price']
price_column = pd.DataFrame(price)
price_column["Price"] = price_column["Price"].str.replace('$', '')
price_column["Price"] = price_column["Price"].str.replace(',', '')
price_column.head(10)
```

```python
#Exract the deal_rating column and split the string to remove the characters.
deal_rating = df2['Deal_Rating']
deal_rating_column = pd.DataFrame(deal_rating.str.split(expand = True)).drop(columns=[1, 2, 4])
deal_rating_column.columns = ['Rate', 'Under_Value($)']
deal_rating_column["Under_Value($)"] = deal_rating_column["Under_Value($)"].str.replace('$', '')
deal_rating_column["Under_Value($)"] = deal_rating_column["Under_Value($)"].str.replace(',', '')
deal_rating_column['Under_Value($)'] = deal_rating_column['Under_Value($)'].fillna(0)

#Convert the object type to numeric
deal_rating_column['Under_Value($)'] = pd.to_numeric(deal_rating_column['Under_Value($)'])
deal_rating_column.loc[deal_rating_column["Rate"] == "Home", "Rate"] = "Home_Delivery"
deal_rating_column.head(10)
```

```python
# Remove the null values and convert the object type to numeric
price_column = price_column[price_column.Price != 'Not Priced']
mileage_column = mileage_column[mileage_column.Mileage != 'n/a']
deal_rating_column = deal_rating_column[deal_rating_column.Rate != 'n/a']
price_column['Price'] = pd.to_numeric(price_column['Price'])
mileage_column['Mileage'] = pd.to_numeric(mileage_column['Mileage'])
```

```python
# Remove the old columns and replace with the new columns
df2_removed = df2.drop(columns=['Mileage','Price','Deal_Rating'])
df3 = pd.concat([df2_removed, mileage_column, price_column, deal_rating_column], axis=1, join='inner')
df3.head(30)
```

```python
df3.info()
```

### Clean the Distance Column

```python
#adding new columns for string split. need to get miles by itself
df3[['miles','mile label','from','zip_code']]=df3['Distance'].str.split(" ",expand=True)
df3.head(10)
```

```python
#making initial dataframe for electronic dealers and saying any column with miles=='n/a', 
# then make electronic dealer equal 1.
df3['electronic_dealer'] = 0
df3.loc[df3.miles=='n/a','electronic_dealer'] = 1
```

```python
#make miles column 0 instead of n/a. better for consitent data type
df3.loc[df3.miles=='n/a','miles']=0
```

```python
#making miles dataframe equal to 1 if the result of the string split was 'mi.'
df3.loc[df3.miles=='mi.','miles']=1
df3.head()
```

```python
#dropping redundant and unnecessary columns
df3 = df3.drop(columns=['zip_code', 'from','mile label','Distance'])
df3.head(50)
```

```python
# as interger type cleaning
df3['Year'] = df3['Year'].astype(int)
df3['Distance Radius'] = df3['Distance Radius'].astype(int)
df3['Zip'] = df3['Zip'].astype(int)

df3.dtypes
```

```python
# getting rid of string commas so data types can be specified easier for that column, then making integer value
df3['miles'] = df3['miles'].str.replace(',','')
df3 = df3.replace(np.nan,0)
df3['miles'] = df3['miles'].astype(int)

df3.head(50)
```

```python
# Save to the shared folder
df3.to_pickle('/dsa/groups/casestudy2022su/team05/carscom_v02.pkl')
```
